### Q1. Trong số các khách hàng có nhiều hơn 1 đơn hàng (orders), trung vi số ngày giữa 2 lần mua liên tiếp (inter-order gap) xấp xỉ là bao nhiêu

- Phân tích: các bảng cần phân tích: df_orders.
- Cột cần quan tâm: df_orders.customer_id, df_orders.order_date

```python
import pandas as pd

# 1. Lấy các cột cần thiết
orders = df_orders[['customer_id', 'order_date']].copy()

# 2. Chuyển order_date sang datetime
orders['order_date'] = pd.to_datetime(orders['order_date'])

# 3. Sắp xếp theo khách hàng và ngày mua
orders = orders.sort_values(['customer_id', 'order_date'])

# 4. Đếm số đơn của từng khách
customer_order_counts = orders.groupby('customer_id').size()

# 5. Chỉ giữ khách có từ 2 đơn trở lên
multi_order_customers = customer_order_counts[customer_order_counts >= 2].index
orders = orders[orders['customer_id'].isin(multi_order_customers)].copy()

# 6. Lấy ngày mua trước đó của từng khách
orders['prev_order_date'] = orders.groupby('customer_id')['order_date'].shift(1)

# 7. Tính khoảng cách giữa 2 lần mua liên tiếp (đơn vị: ngày)
orders['inter_order_gap_days'] = (
    orders['order_date'] - orders['prev_order_date']
).dt.days

# 8. Chỉ giữ các dòng có gap hợp lệ
gaps = orders.dropna(subset=['inter_order_gap_days']).copy()

# 9. Tính trung vị
median_gap = gaps['inter_order_gap_days'].median()

# 10. Hiển thị kết quả
print(f"Median inter-order gap: {median_gap:.2f} days")

print("\nSample gaps:")
print(
    gaps[['customer_id', 'prev_order_date', 'order_date', 'inter_order_gap_days']]
    .head(10)
)

print("\nSố khách có từ 2 đơn trở lên:", len(multi_order_customers))
print("Số khoảng cách mua hàng hợp lệ:", len(gaps))
```

## Q2. Phân khúc sản phẩm (segment) nào trong products.csv có tỷ suất lợi nhuận gộp trung bình cao nhất, với công thức (price −cogs)/price?
- Phân tích: bảng cần phân tích: df_prod
- Cột cần phân tích df_prod.segment, df_prod.price, df_prod.cogs

```python
import pandas as pd

def avg_ratio_by_segment(df_prod: pd.DataFrame) -> pd.DataFrame:
    prods = df_prod[['segment', 'price', 'cogs']].copy()

    prods = prods.dropna(subset=['segment', 'price', 'cogs'])
    prods = prods[prods['price'] != 0].copy()

    prods['ratio'] = (prods['price'] - prods['cogs']) / prods['price']

    result = (
        prods.groupby('segment', as_index=False)['ratio']
        .mean()
        .sort_values('ratio', ascending=False)
    )

    return result


result = avg_ratio_by_segment(df_prod)
print(result)
```

## Q6. Trong customers.csv, xét các khách hàng có age_group khác null, nhóm tuổi nào có số đơn hàng trung bình trên mỗi khách hàng cao nhất? (tổng số đơn / số khách hàng trong nhóm)

- Bảng cần quan tâm: df_cust, df_orders
- Cột cần quan tâm: df_cust.age_group

```python
import pandas as pd

# 1. Chỉ giữ khách có age_group khác null
cust = df_cust[['customer_id', 'age_group']].copy()
cust = cust[cust['age_group'].notna()].copy()

# 2. Lấy cột customer_id từ orders
orders = df_orders[['customer_id', 'order_id']].copy()

# 3. Đếm số đơn theo từng khách
orders_per_customer = (
    orders.groupby('customer_id')['order_id']
    .count()
    .reset_index(name='total_orders')
)

# 4. Ghép với bảng khách hàng
cust_orders = cust.merge(orders_per_customer, on='customer_id', how='left')

# 5. Khách không có đơn thì gán 0
cust_orders['total_orders'] = cust_orders['total_orders'].fillna(0)

# 6. Tính theo từng nhóm tuổi
age_group_stats = (
    cust_orders.groupby('age_group')
    .agg(
        total_orders=('total_orders', 'sum'),
        total_customers=('customer_id', 'nunique')
    )
    .reset_index()
)

print(age_group_stats)
age_group_stats['avg_orders_per_customer'] = (
    age_group_stats['total_orders'] / age_group_stats['total_customers']
)

# 7. Sắp xếp giảm dần để lấy nhóm cao nhất
age_group_stats = age_group_stats.sort_values(
    'avg_orders_per_customer', ascending=False
)

print(age_group_stats)

# 8. Nhóm tuổi cao nhất
top_group = age_group_stats.iloc[0]

print("\nNhóm tuổi có số đơn hàng trung bình trên mỗi khách hàng cao nhất:")
print(f"age_group: {top_group['age_group']}")
print(f"avg_orders_per_customer: {top_group['avg_orders_per_customer']:.2f}")
```

##  Q7. Vùng (region) nào trong geography.csv tạo ra tổng doanh thu cao nhất trong sales_train.csv

```python
import pandas as pd

# 1. Các bảng cần thiết
orders = df_orders[['order_id', 'zip']].copy()
payments = df_payments[['order_id', 'payment_value']].copy()
geo = df_geo[['zip', 'region']].copy()

# 2. Tổng tiền refund theo order
refunds = (
    df_returns.groupby('order_id', as_index=False)['refund_amount']
    .sum()
    .rename(columns={'refund_amount': 'total_refund'})
)

# 3. Gộp payment với refund
df = payments.merge(refunds, on='order_id', how='left')
df['total_refund'] = df['total_refund'].fillna(0)

# 4. Tính doanh thu thuần
df['net_revenue'] = df['payment_value'] - df['total_refund']

# 5. Gắn zip và region
df = df.merge(orders, on='order_id', how='left')
df = df.merge(geo, on='zip', how='left')

# 6. Tổng doanh thu thuần theo vùng
region_revenue = (
    df.groupby('region', as_index=False)['net_revenue']
    .sum()
    .sort_values('net_revenue', ascending=False)
)

print(region_revenue)

top_region = region_revenue.iloc[0]
print("\nVùng có doanh thu thuần cao nhất:")
print(f"Region: {top_region['region']}")
print(f"Net revenue: {top_region['net_revenue']:.2f}")
```

## Q9. Trong bốn kích thước sản phẩm (S, M, L, XL), kích thước nào có tỷ lệ trả hàng cao nhất, được định nghĩa là số bản ghi trong returns chia cho số dòng trong order_items (join với products theo product_id)?

```python
import pandas as pd

# 1. Lấy product_id và size
products = df_prod[['product_id', 'size']].copy()

# 2. Join order_items với products để gắn size
order_items_with_size = df_order_items.merge(
    products,
    on='product_id',
    how='left'
)

# 3. Join returns với products để gắn size
returns_with_size = df_returns.merge(
    products,
    on='product_id',
    how='left'
)

# 4. Đếm số dòng order_items theo size
order_item_counts = (
    order_items_with_size.groupby('size')
    .size()
    .reset_index(name='order_item_count')
)

# 5. Đếm số bản ghi returns theo size
return_counts = (
    returns_with_size.groupby('size')
    .size()
    .reset_index(name='return_count')
)

# 6. Gộp 2 bảng lại
size_return_rate = order_item_counts.merge(
    return_counts,
    on='size',
    how='left'
)

# 7. Size nào không có return thì gán 0
size_return_rate['return_count'] = size_return_rate['return_count'].fillna(0)

# 8. Tính tỷ lệ trả hàng
size_return_rate['return_rate'] = (
    size_return_rate['return_count'] / size_return_rate['order_item_count']
)

# 9. Sắp xếp giảm dần
size_return_rate = size_return_rate.sort_values(
    'return_rate',
    ascending=False
).reset_index(drop=True)

print(size_return_rate)

# 10. Lấy size có tỷ lệ trả hàng cao nhất
top_size = size_return_rate.iloc[0]

print("\nKích thước có tỷ lệ trả hàng cao nhất:")
print(f"Size: {top_size['size']}")
print(f"Return rate: {top_size['return_rate']:.4f}")
print(f"Return count: {int(top_size['return_count'])}")
print(f"Order item count: {int(top_size['order_item_count'])}")
```

## Q10. Trong payments.csv, kế hoạch trả góp nào có giá trị thanh toán trung bình trên mỗi đơn hàng cao nhất?

```python
import pandas as pd

# 1. Lấy 2 cột installments và payment_value
payments = df_payments[['installments', 'payment_value']].copy()

# 2. Tính giá trị thanh toán trung bình theo từng kế hoạch trả góp
installment_stats = (
    payments.groupby('installments', as_index=False)
    .agg(avg_payment_value=('payment_value', 'mean'))
)

# 3. Sắp xếp giảm dần để thống kê và tìm nhóm cao nhất
installment_stats = installment_stats.sort_values(
    'avg_payment_value', ascending=False
).reset_index(drop=True)

print(installment_stats)

# Lấy kế hoạch trả góp có avg cao nhất
top_plan = installment_stats.iloc[0]

print("\nKế hoạch trả góp có giá trị thanh toán trung bình cao nhất:")
print(f"Installments: {int(top_plan['installments'])}")
print(f"Average payment value: {top_plan['avg_payment_value']:.2f}")
```
